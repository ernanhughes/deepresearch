```python
from smolagents import CodeAgent, LiteLLMModel

model_id = "ollama/llama3.2"
model = LiteLLMModel(model_id=model_id, api_base="http://localhost:11434")

agent = CodeAgent(tools=[], model=model, additional_authorized_imports=['requests', 'bs4', 'web'])
agent.run("Could you get me the title of the page at url 'https://programmer.ie'?")
```

```python
from smolagents import CodeAgent, DuckDuckGoSearchTool, LiteLLMModel, VisitWebpageTool

from smolagents.agents import (
    CodeAgent,
    ToolCallingAgent,
)

model_id = "ollama/qwen2.5-coder:14b"
model = LiteLLMModel(model_id=model_id, api_base="http://localhost:11434")

search_agent = CodeAgent(
    tools=[DuckDuckGoSearchTool(), VisitWebpageTool()],
    model=model,
    name="search_agent",
    description="This is an agent that can do web search.",
)

manager_agent = CodeAgent(
    tools=[],
    model=model,
    managed_agents=[search_agent],
)
manager_agent.run(
    "If the US keeps its 2024 growth rate, how many years will it take for the GDP to double?"
)
```
